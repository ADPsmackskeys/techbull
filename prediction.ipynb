```python
import yfinance as yf
import datetime
import plotly.graph_objects as go
import pandas as pd

data = yf.download ('TCS.NS', start = '2024-07-07', end = datetime.date.today())
# print (type(data))
# data.to_csv ('TCS_data1.csv')
data = pd.read_csv ('TCS_data1.csv')
l = tuple (data)
print (data)
chart = go.Figure (data = go.Candlestick(x = data[l[0]], open = data['Open'], high = data['High'], low = data['Low'], close = data['Close']))
chart.show()
```

## Introduction to Japanese Candlesticks and Candlestick plots
The graph shown above depicts the **Open price(O), High price(H), Low price(L), Close price(C)** in the form of **candles**.

Candlesticks consist of two main parts: <br>
    1. **The Real Body**: The range between O and C <br>
    2. **The Wicks**: The differences between H & C and O & L, for a Bullish candle.

Candlesticks can be of 2 types: **Bullish** or **Bearish**

```python
from sklearn.preprocessing import *
le = LabelEncoder ()
data ['Date'] = le.fit_transform (data['Date'])
data.head
```

``LabelEncoder ()`` performs Label Encoding. It is a technique that is used to convert categorical columns into numerical ones so that they can be fitted by machine learning models which only take numerical data. 

``fit_transform ()`` learns the parameters and apply the transformation to new data.

```python
pr_dat = input ('Enter the data to predict: ')
x = data.drop (columns = [pr_dat], axis = 1)
y = data [pr_dat]
x
```

Now, we need to **specify which value (column)** we need to predict. In our case, we will be predicting the **Close** price.

```python
scale = StandardScaler ()
names = x.columns
x = scale.fit_transform (x)
x = pd.DataFrame (x, columns = names)
x
```

``StandardScaler ()`` is used to standardize features by removing the mean and scaling to unit variance.
The standard score of a sample x is calculated as:

    z = (x - u) / s

where **u is the mean** of the training samples or zero if with_mean = False, and **s is the standard deviation** of the training samples or one if with_std = False.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2, random_state = 0)
from sklearn.linear_model import LinearRegression
model = LinearRegression ()
model.fit (x_train, y_train)
```

We are using **Linear Regression** to predict the values of the Closing Price of each day.
We will be comparing the predicted values with the actual values of the day to find out the accuracy.

``test_size = 0.2`` means that we are using **80% of the data to _train_ the model**, and **20% of the data to _test_ the model**.

```python
pred = model.predict(x_test)
pred
```

```python
y_test
```

```python
E = pred - y_test
E
```

```python
from sklearn.metrics import r2_score
r2_score (pred, y_test)*100
```

The R<sup>2</sup> score of a Linear Regression Model is the proportion of the variation in the dependent variable that is predictable from the independent variable.

R<sup>2</sup> = 1: Perfect fit. The model explains all the variability in the target variable.

R<sup>2</sup> = 0: The model does not explain any variability in the target variable. 

0 < R<sup>2</sup> < 1: The model explains a proportion of the variability in the target variable.
